```julia
using Plots, LinearAlgebra
```

```julia
plotlyjs()

include("boids.jl")

width = height = 1000
nb_boids = 200
alignValue = 0.5
separationValue = 1
cohesionValue = 0.6

flock = [Boid(i, rand(1:width, 3), randn(3).*2, randn(3)) for i=1:nb_boids];
```

```julia
anim = @animate for i in 1:1000
    for boid in flock
        # update boids with the corresponding slider values
        fly!(boid, flock, alignValue[], separationValue[], cohesionValue[])
    end
    scatter(map(x -> x.position[1], flock), map(x -> x.position[2], flock), map(x -> x.position[3], flock), lims=(1,1000));
end

```

```julia
gif(anim, "anim.gif", fps = 30)
```
